```python
import csv
import matplotlib.pyplot as plt

f=open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)

mx = ['0'] * 23
mx_station = [''] * 23

for row in data :
    row[4:] = map(int, row[4:])
    for j in range(23) :
        a = row[j*2+4]
        if a > int(mx[j]) :
            mx[j] = a
            mx_station[j] = row[3] + '(' + str(j+4) + ')'
            
plt.rc('font', family='Malgun Gothic')
plt.bar(range(23), mx)
plt.xticks(range(23), mx_station, rotation=90)
plt.xlabel('역이름')
plt.ylabel('인구수')
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
import numpy as np

f = open('pop202306.csv')
data = csv.reader(f)
next(data)

name1 = input('지역이름1을 입력 : ')
name2 = input('지역이름2을 입력 : ')

for row in data:
    if name1 in row[0] :
        for i in row[3:] :
            home1 = np.array(row[3:], dtype=int)
        break

for row in data:
    if name2 in row[0] :
        for i in row[3:] :
            home2 = np.array(row[3:], dtype=int)
        break

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=-False
plt.title(name1 + name2 + '지역의 인구구조')
plt.plot(home1)
plt.plot(home2)
plt.show() 
```
