```python
import csv
import matplotlib.pyplot as plt

f = open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)

mx = ['0'] * 23
mx_station = [''] * 23

for row in data :
    row[4:] = map(int, row[4:])
    for j in range(23) :
        a = row[j*2+4]
        if a > int(mx[j]) :
            mx[j] = a
            mx_station[j] = row[3]+'(' + str(j+4) + ')'
            
plt.rc('font', family='Malgun Gothic')
plt.bar(range(23), mx)
plt.xticks(range(23), mx_station ,rotation=90)
plt.show()
```

```python
import csv
f = open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)
s_in = [0] * 24
s_out = [0] * 24

for row in data :
    row[4:] = map(int, row[4:])
    for i in range(24):
        s_in[i] += row[4+i*2]
        s_out[i] += row[5+i*2]

plt.rc('font', family='Malgun Gothic')
plt.title('   ')
plt.plot(s_in, label='')
plt.plot(s_out, label='')
plt.legend()
plt.xticks(range(24), range(4, 28))
plt.show()
```
