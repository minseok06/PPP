```python
from wordcloud import WordCloud
from string import digits
import re
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

font_path = 'C:/Windows/Fonts/malgun.ttf' # 글꼴 지정
plt.rc('font',family='Malgun Gothic') # 그래프 한글 패치
```

# 이미지는 수학 배열이다

```python
text = ''
with open("카톡.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()
    for line in lines:
        text += line
        
text = re.sub(r'[0-9]+','',text)

wc = WordCloud(font_path=font_path, background_color='white',
               width=600, height=400)

wc.generate(text)
plt.figure(figsize=(8,8))
plt.imshow(wc)
plt.title('2-9반', size=30)
plt.axis('off')
plt.show()
```

```python
text = ''
with open('교복 하면 떠오르는 것.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    for line in lines:
        text += line
        
text = re.sub(r'[0-9]+','',text)

wc = WordCloud(font_path=font_path, background_color='white',
               prefer_horizontal=True, width=600, height=400)

wc.generate(text)
plt.figure(figsize=(8,8))
plt.imshow(wc)
plt.title('교복하면 생각나는 단어들', size=30)
plt.axis('off')
plt.show()
```
