```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
k = pd.read_csv('교복.csv', encoding='cp949')
k.head()
```

```python
k['교복 입기 싫은 날'].value_counts()
```

```python
sns.set_palette('Set2')
sns.countplot(data=k, x='교복 입기 싫은 날', order=['월','화','수','목','금','없다'])
```

```python
sns.countplot(data=k, x='입는 이유')
plt.xticks(rotation=20)
```

```python
sns.countplot(data=k, x='안 입는 이유')
# plt.xticks(rotation=20)
plt.title('교복을 안 입는 이유', fontsize=20)
```

```python
k['교복 단어'].value_counts().plot(kind='bar')
plt.rc('font', family = 'Malgun Gothic')
plt.xticks(rotation=90)
plt.show()
```
