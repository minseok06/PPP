# 우리동네 인구분포 시각화 하기

```python
import csv
import matplotlib.pyplot as plt

f = open('202306_202306_연령별인구현황_월간.csv')
data = csv.reader(f)

result = []

for row in data:
    if '금광1동' in row[0]:
        for i in row[3:]:
            result.append(int(i))

plt.rc('font', family='Malgun Gothic')
plt.title('금광1동 지역의 인구분포')
plt.plot(result, 'skyblue', label='연령별분포')
plt.legend()
plt.show()
```
