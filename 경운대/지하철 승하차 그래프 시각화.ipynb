# 아침 7~9시 60만명 이상 승차하는 역

```python
import csv
import matplotlib.pyplot as plt

f = open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)

for row in data :
    row[4:] = map(int, row[4:])
    if sum(row[10:15:2]) > 600000:
        mx =  sum(row[10:15:2])
        mx_station = row[3] + '('+row[1]+') : '
        print(mx_station, mx)
```

# 시간대별 30만명 이상 승차하는 역&인원 출력

```python
import csv
import matplotlib.pyplot as plt

f = open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)

t1 = input('지하철역 승하차 시작시간을 입력해주세요. : ')
t2 = input('지하철역 승하차 마지막시간을 입력해주세요. : ')
print('------------------------------------------------')
print('{} ~ {}시 사이에 30만명 이상 승차하는 역 입니다.'.format(t1,t2))
print('------------------------------------------------')

a = 2*int(t1)-4
b = 2*int(t2)-4
# -4이면 승차
# -3이면 하차

for row in data :
    row[4:] = map(int, row[4:])
    if sum(row[a:b:2]) > 300000:
        mx = sum(row[a:b:2])
        mx_station = row[3] + '(' +row[1]+') :'
        print(mx_station, mx)
```

# 각 시간대별 최대 승차역 및 인원 출력

```python
import csv
f = open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)
mx = [0] * 24
mx_station = [''] * 24

for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24) :
        a = row[j*2+4]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3]
print(mx_station)
print(mx)
```

```python
import csv
import matplotlib.pyplot as plt

f = open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)

mx = 0
mx_station = ''

t = int(input('몇 시의 승차 인원이 가장 많은 역이 굼금합니까? : '))

for row in data:
    row[4:] = map(int, row[4:])
    a = row[t*2-4]
    if a > mx :
        mx = a
        mx_station = row[3] + '(' + row[1] + ') : '
print(mx_station, mx)
```

# 시간대별 최대 승차역 및 인원 막대 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)

mx = ['0'] * 23
mx_station = [''] * 23

for row in data :
    row[4:] = map(int, row[4:])
    for j in range(23) :
        a = row[j*2+4]
        if a > int(mx[j]):
            mx[j] = a
            mx_station[j] = row[3]
            
plt.rc('font', family='Malgun Gothic')
plt.bar(range(23), mx)
plt.xticks(range(23), mx_station, rotation=90)
plt.show()
```

# 시간대 추가하기 + '('+str(j+4)+')'

```python
import csv
import matplotlib.pyplot as plt

f = open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)

mx = ['0'] * 23
mx_station = [''] * 23

for row in data :
    row[4:] = map(int, row[4:])
    for j in range(23) :
        a = row[j*2+4]
        if a > int(mx[j]):
            mx[j] = a
            mx_station[j] = row[3] + '('+str(j+4)+')'
            
plt.title('시간대별 승차인원 최고역')
plt.rc('font', family='Malgun Gothic')
plt.bar(range(23), mx, color='hotpink')
plt.xticks(range(23), mx_station, rotation=90)
plt.show()
```

# 시간대별 최대 하차역 및 인원 막대 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt
f = open('subway202306.csv')
data = csv.reader(f)
next(data)
next(data)

mx = [0] * 24
mx_station = [''] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        a = row[j*2+5]
        if a > mx[j]:
            mx[j] = a
            mx_station[j] = row[3] + '('+str(j+4)+'시)'
            
plt.title('시간대별 하차인원 최고역')
plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx, color='skyblue')
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```
