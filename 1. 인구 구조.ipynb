```python
import csv
import matplotlib.pyplot as plt

f = open('202306_202306_연령별인구현황_월간.csv')
data = csv.reader(f)

result = []
name = input('지역을 입력해주세요 : ')

for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))

# plt.plot(result)
plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.title(name+' 연령별 인구분포')
plt.plot(result, 'skyblue', label='연령별분포')
plt.legend()
plt.show()
```
