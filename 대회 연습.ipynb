```python
import csv
f = open('metro.csv')
data = csv.reader(f)
next(data)
next(data)
result = []
for row in data:
    row[4:] = map(int,row[4:])
    result.append(row[10])
print(len(result))
print(result)
```

```python
import matplotlib.pyplot as plt
plt.bar(range(len(result)), result)
plt.show()
```

```python
import csv
f = open('metro.csv')
data = csv.reader(f)
next(data)
next(data)

mx = 0
mx_station = ''

for row in data:
    row[4:] = map(int, row[4:])
    if sum(row[11:16:2]) > mx:
        mx = sum(row[11:16:2])
        mx_station = row[3]+'('+row[1]+')'
        print(mx_station, mx)
print('----------------------------------')
print()
print(mx_station, mx)
```
