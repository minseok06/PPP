```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 성일정보고등학교 학생 데이터
# 변수명 => SHD

```python
SHD = pd.read_csv('data/1급-기밀파일.csv', encoding='cp949')
SHD.head(1)
```

# 1,2,3학년 별 설문 참여도 (1학년 최고)

```python
SHD['학년'].value_counts()
# 학년별 설문조사 참여 수
```

```python
plt.rc('font',family='Malgun Gothic')

plt.figure(figsize=(10,6))
explode=[0.02,0.04,0.03]
colors = ['#8BC34A','#5C6BC0','#EF5350']
labels=['1학년','2학년','3학년']
plt.pie(SHD['학년'].value_counts(),
        labels=labels,
        autopct='%.1f%%',
        counterclock=False,
        startangle=90,
        explode=explode,
        colors=colors,
       )
plt.title('학년 별 설문 참여 비율', fontsize=15)
plt.legend(loc=(1, 0.7),title = '학년')
plt.show()
```

# 7개 학과 별 설문 참여도 (전산계열이 압도적)

```python
SHD['학과'].value_counts()
# 학과별 설문조사 참여 수
```

```python
plt.rc('font',family='Malgun Gothic')
sns.set_palette("Set3")

plt.figure(figsize=(10,6))
explode=[0.02,0.04,0.03,0.02,0.04,0.03,0.02]
labels=['소프트웨어개발과','스마트웹콘텐츠과','창업마케팅과',
        '부사관과','금융경영과','뷰티디자인과','회계정보과']
plt.pie(SHD['학과'].value_counts(),
        labels=labels,
        autopct='%.1f%%',
        counterclock=False,
        startangle=90,
        explode=explode,
       )
plt.title('학과 별 설문 참여 비율', fontsize=15)
plt.legend(loc=(1, 0.7),title = '학년')
plt.show()
```

# 학과별에 따른 학년별 설문 참여도

# 학년별로 추출해 SHD_0 형식으로 변수 저장

```python
SHD_1 = SHD[SHD['학년']=='1학년'] # 1학년 데이터만 추출(변수 저장)
SHD_2 = SHD[SHD['학년']=='2학년'] # 2학년 데이터만 추출(변수 저장)
SHD_3 = SHD[SHD['학년']=='3학년'] # 3학년 데이터만 추출(변수 저장)
```

# 알바 여부 비율이 가장 높은 학년은?

```python
sns.set_palette("pastel")
SHD.groupby(['학년'])['알바'].value_counts().plot(kind='bar') #학년 그룹 알바 컬럼 나열
plt.rc('font',family='Malgun Gothic') # 텍스트 한글 패치
plt.title('학년 별 알바 여부', fontsize=25) # 제목 텍스트 크기
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

# 전학년 소셜미디어 사용시간

```python
sns.set_palette("Set3")
SHD['소셜미디어 사용시간'].value_counts().plot(kind='bar')
plt.title('소셜미디어 사용시간', fontsize=20) # 제목 텍스트 크기
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
sns.set_palette("Set3")
sns.countplot(data=SHD, x='소셜미디어 사용시간', hue='학년', order=['2시간-3시간 이상','1시간-2시간','30-1시간','0-30분'])
plt.title('학년 별 소셜미디어 사용시간', fontsize=20)
plt.show()
```

# 성일정보고등학교 전공 선호도는
- 1학년은 평균 6점
- 2학년은 평균 7점
- 3학년은 평균 5점

# 평균적으로 10점 만점 중 6점이다.

```python
sns.set_palette("pastel")
sns.boxplot(data=SHD, x='학년', y='전공 선호도', order=['1학년','2학년','3학년'])
plt.title('학년 별 전공선호도 평균')
# 박스 안에 있는 선은 평균을 나타냅니다.
plt.show()
```

# 성일정보고 학생들의 취미 1위는 게임

# new_SHD 변수에 데이터가 동등하게
# 남자를 랜덤으로 88명 뽑아주었다.

```python
male = SHD[SHD['성별']=='남자'].sample(n = 88)
female = SHD[SHD['성별']=='여자']
```

# 전

```python
SHD['성별'].value_counts()
```

# 후

```python
new_SHD = pd.concat([male,female])
new_SHD['성별'].value_counts()
```

```python
sns.set_palette("Set2")
sns.countplot(data=new_SHD, x='성별', hue='취미')
plt.title('우리학교 학생 성별에 따른 취미', fontsize=20)
plt.show()
```

```python
sns.set_palette("pastel")
sns.countplot(data=new_SHD, x='취미', hue='성별')
plt.title('게임 1위', fontsize=20)
plt.show()
```
